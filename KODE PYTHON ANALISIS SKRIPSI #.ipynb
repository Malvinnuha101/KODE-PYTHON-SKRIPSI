```python
#!/usr/bin/env python3  # Shebang untuk menjalankan skrip sebagai executable Python3, memastikan kompatibilitas lingkungan.
# -*- coding: utf-8 -*-  # Encoding UTF-8 untuk mendukung karakter khusus seperti simbol matematika.

import numpy as np  # Impor NumPy untuk operasi array numerik efisien, dasar perhitungan vektor.
import matplotlib.pyplot as plt  # Impor Matplotlib untuk plotting grafik, visualisasi data rotasi.
from scipy.optimize import curve_fit, root_scalar  # Impor curve_fit untuk fitting model ke data, root_scalar untuk solving persamaan.
from scipy.integrate import quad  # Impor quad untuk integrasi numerik, digunakan dalam perhitungan densitas.
from scipy.special import iv, kv  # Impor fungsi Bessel modified untuk model disk galaksi.
from functools import lru_cache  # Impor lru_cache untuk memoization, mengoptimalkan fungsi rekursif/berulang.
from sklearn.metrics import r2_score  # Impor R² score untuk evaluasi kualitas fitting model.
import scipy.stats as stats  # Impor modul statistik SciPy untuk uji seperti Shapiro-Wilk.
import warnings  # Impor warnings untuk mengelola peringatan runtime.

warnings.filterwarnings('ignore', category=RuntimeWarning)  # Abaikan RuntimeWarning untuk menghindari output berlebih pada NaN/Inf.
warnings.filterwarnings('ignore', category=UserWarning)  # Abaikan UserWarning, misalnya dari fitting yang tidak konvergen sempurna.

G_ASTRO = 4.30091e-6  # Konstanta gravitasi dalam satuan astronomi (km²/s²/kpc/M_sun), untuk perhitungan kecepatan orbital.
A0_STD_SI = 1.2e-10  # Nilai a0 standar MOND dalam SI (m/s²), percepatan kritis di regime MOND.
A0_ARCTAN_SI = A0_STD_SI  # Set a0 untuk model arctan sama dengan standar, untuk konsistensi.
KM2_S2_PER_KPC_TO_SI = 3.2408e-14  # Faktor konversi satuan dari km²/s²/kpc ke SI, untuk akurasi.
A0_STD_ASTRO = A0_STD_SI / KM2_S2_PER_KPC_TO_SI  # Konversi a0 standar ke satuan astro.
A0_ARCTAN_ASTRO = A0_ARCTAN_SI / KM2_S2_PER_KPC_TO_SI  # Konversi a0 arctan ke satuan astro.
KAPPA = 7.6695  # Konstanta kappa untuk profil bulge de Vaucouleurs, dari fitting empiris.

class GalaxyRotationAnalyzer:  # Definisi kelas utama untuk menganalisis rotasi galaksi, mengenkapsulasi data dan metode.
    def __init__(self):  # Inisialisasi objek kelas.
        self.data = self.load_observational_data()  # Muat data observasi untuk Bima Sakti dan M31.
        self.models = {  # Dictionary model: kunci nama model, nilai fungsi kecepatan total.
            'MOND Std': self.v_total_mond_std,  # Model MOND standar.
            'MOND Arctan': self.v_total_mond_arctan,  # Model MOND dengan interpolasi arctan.
            'ΛCDM NFW': self.v_total_nfw,  # Model ΛCDM dengan halo NFW.
            'ΛCDM Core': self.v_total_core  # Model ΛCDM dengan halo core.
        }
        self.initial_params = self.set_initial_parameters()  # Set parameter awal dan batas untuk fitting.

    @staticmethod  # Metode statis, tidak bergantung pada instance.
    def load_observational_data():  # Fungsi untuk memuat data radius (r), kecepatan (v), dan error (sigma_v) dari observasi.
        r_mw = np.concatenate([  # Gabung array radius untuk Bima Sakti (Milky Way), dalam kpc.
            np.array([0.100, 0.110, 0.123, 0.133, 0.146, 0.161, 0.177, 0.195, 0.214, 0.236,
                      0.258, 0.285, 0.314, 0.345, 0.380, 0.418, 0.459, 0.505, 0.556, 0.612,
                      0.673, 0.741, 0.814, 0.895, 0.985, 1.083, 1.192, 1.311, 1.442, 1.586,
                      1.745, 1.919, 2.111, 2.323, 2.555, 2.811, 3.091]),
            np.array([3.400, 3.740, 4.114, 4.526, 4.979, 5.476, 6.024, 6.626, 7.289, 8.018,
                      8.820, 9.702, 10.672, 11.739, 12.913, 14.204, 15.625, 17.187, 18.906,
                      20.797, 22.876, 25.164, 27.680, 30.448, 33.493, 36.842, 40.527, 44.579,
                      49.037, 53.941, 59.335, 65.268, 71.795, 78.975, 86.872, 95.561])
        ])
        v_mw = np.concatenate([  # Gabung array kecepatan rotasi untuk Bima Sakti, dalam km/s.
            np.array([144.9, 147.4, 150.4, 153.8, 158.9, 162.4, 180.1, 196.6, 213.6, 227.8,
                      234.4, 244.1, 248.2, 250.2, 251.0, 250.7, 249.7, 248.0, 245.0, 243.2,
                      239.8, 235.3, 231.7, 227.8, 224.5, 221.7, 210.1, 216.8, 214.7, 212.7,
                      211.2, 209.5, 208.5, 208.2, 208.9, 210.4, 213.4]),
            np.array([217.2, 222.0, 226.6, 229.5, 231.6, 234.1, 237.2, 240.5, 241.1, 236.0,
                      236.7, 234.5, 234.2, 237.1, 242.8, 248.5, 240.7, 246.2, 246.5, 243.9,
                      241.6, 243.7, 237.3, 229.6, 222.5, 218.0, 207.1, 201.5, 194.2, 189.8,
                      186.2, 184.7, 183.9, 181.4, 179.5, 167.7])
        ])
        sigma_v_mw = np.concatenate([  # Gabung array error kecepatan untuk Bima Sakti.
            np.array([3.7, 4.2, 4.8, 6.1, 10.3, 16.1, 23.4, 27.1, 26.9, 22.7, 17.0,
                      11.8, 7.6, 4.7, 2.9, 2.1, 2.3, 2.9, 3.7, 4.6, 5.4, 6.7, 6.5,
                      6.0, 5.2, 4.5, 4.0, 3.7, 3.4, 3.1, 2.9, 2.3, 1.8, 1.6, 2.2, 3.6, 4.8]),
            np.array([5.9, 6.6, 5.7, 4.1, 4.3, 5.3, 5.7, 5.0, 4.1, 4.4, 5.1, 6.0,
                      7.1, 9.8, 12.4, 13.3, 14.8, 17.4, 17.3, 17.5, 15.6, 15.2, 16.1,
                      15.3, 14.1, 14.0, 13.8, 12.7, 12.9, 11.1, 10.4, 9.6, 9.3, 13.0, 14.6, 16.5])
        ])

        r_m31 = np.concatenate([  # Gabung array radius untuk M31 (Andromeda).
            np.array([0.100, 0.120, 0.144, 0.173, 0.207, 0.249, 0.299, 0.358, 0.430, 0.516,
                      0.619, 0.743, 0.892, 1.070, 1.284, 1.541, 1.849, 2.219, 2.662, 3.195,
                      3.834, 4.601, 5.521, 6.625, 7.950, 9.540, 11.448, 13.737]),
            np.array([16.5, 19.8, 23.7, 28.5, 34.2, 41.0, 49.2, 59.1, 70.9, 85.1,
                      102.1, 122.5, 147.0, 176.4, 211.6, 254.0, 304.8, 365.7, 438.9, 526.6,
                      632.0, 758.4, 910.0, 1092.1, 1310.5, 1572.6, 1887.1])
        ])
        v_m31 = np.concatenate([  # Gabung array kecepatan untuk M31.
            np.array([183.9, 190.7, 204.1, 207.6, 210.2, 213.5, 217.2, 219.7, 219.4, 216.2,
                      213.7, 219.5, 221.3, 235.3, 242.8, 248.3, 252.9, 245.0, 239.0, 236.6,
                      228.4, 222.7, 229.1, 243.2, 255.4, 261.4, 263.5, 262.8]),
            np.array([257.3, 246.1, 236.5, 232.3, 233.5, 230.4, 237.1, 249.4, 218.5, 206.6,
                      213.0, 197.4, 177.5, 165.3, 165.1, 160.8, 151.1, 122.0, 168.3, 200.5,
                      239.0, 248.1, 250.4, 241.9, 235.0, 332.7, 414.9])
        ])
        sigma_v_m31 = np.concatenate([  # Gabung array error kecepatan untuk M31.
            np.array([15.0, 10.3, 22.6, 4.6, 3.9, 4.2, 3.8, 2.2, 2.2, 3.7, 10.9, 22.6,
                      30.7, 55.7, 57.0, 52.4, 50.6, 35.2, 16.5, 9.2, 21.1, 30.4, 29.0,
                      23.6, 18.5, 14.7, 11.4, 10.0]),
            np.array([12.2, 13.4, 11.7, 16.6, 27.4, 52.4, 109.4, 134.6, 132.4, 135.0,
                      132.5, 105.7, 81.8, 74.7, 76.9, 77.4, 74.8, 96.5, 125.7, 123.4,
                      147.8, 145.0, 141.4, 138.7, 103.5, 109.5, 69.6])
        ])

        return {  # Kembalikan dictionary data untuk kedua galaksi.
            'Bima Sakti': {'r': r_mw, 'v': v_mw, 'sigma_v': sigma_v_mw},  # Data Bima Sakti.
            'M31': {'r': r_m31, 'v': v_m31, 'sigma_v': sigma_v_m31}  # Data M31.
        }

    @staticmethod  # Metode statis.
    def sigma_b(r, Sigma_be, a_b):  # Fungsi densitas permukaan bulge (de Vaucouleurs profile).
        return Sigma_be * np.exp(-KAPPA * ((r / a_b)**0.25 - 1))  # Hitung sigma bulge eksponensial.

    @staticmethod  # Metode statis.
    def dsigma_b_dx(x, Sigma_be, a_b):  # Turunan densitas bulge terhadap x (untuk integrasi).
        return GalaxyRotationAnalyzer.sigma_b(x, Sigma_be, a_b) * (-KAPPA / (4 * a_b)) * (x / a_b)**(-0.75)  # Hitung derivatif.

    @lru_cache(maxsize=None)  # Cache untuk optimasi, maxsize tak terbatas.
    def rho_b_scalar(self, r_val, Sigma_be, a_b):  # Densitas volume bulge untuk radius tunggal.
        if r_val == 0:  # Kasus khusus r=0 untuk menghindari singularitas.
            integrand = lambda x: self.dsigma_b_dx(x, Sigma_be, a_b) / x  # Integrand untuk r=0.
            integral, _ = quad(integrand, 1e-10, np.inf, epsabs=1e-4, epsrel=1e-4, limit=100)  # Integrasi numerik.
        else:  # Kasus r > 0.
            integrand = lambda x: self.dsigma_b_dx(x, Sigma_be, a_b) / np.sqrt(x**2 - r_val**2)  # Integrand Poisson equation.
            integral, _ = quad(integrand, r_val, np.inf, epsabs=1e-4, epsrel=1e-4, limit=100)  # Integrasi.
        return -1 / np.pi * integral  # Kembalikan rho (densitas) negatif dari integral (konvensi).

    def rho_b(self, r, Sigma_be, a_b):  # Densitas volume bulge untuk array radius.
        return np.array([self.rho_b_scalar(ri, Sigma_be, a_b) for ri in np.atleast_1d(r)])  # Loop atas array r.

    @lru_cache(maxsize=None)  # Cache.
    def M_b_scalar(self, r_val, Sigma_be, a_b):  # Massa bulge dalam radius tunggal.
        if r_val == 0:  # Kasus r=0, massa=0.
            return 0.0
        integrand = lambda s: 4 * np.pi * s**2 * self.rho_b_scalar(s, Sigma_be, a_b)  # Integrand massa spherical.
        mass, _ = quad(integrand, 1e-10, r_val, epsabs=1e-4, epsrel=1e-4, limit=100)  # Integrasi.
        return mass  # Kembalikan massa.

    def M_b(self, r, Sigma_be, a_b):  # Massa bulge untuk array radius.
        return np.array([self.M_b_scalar(ri, Sigma_be, a_b) for ri in np.atleast_1d(r)])  # Loop atas array r.

    def v_bulge_sq(self, r, Sigma_be, a_b):  # Kuadrat kecepatan bulge.
        r_arr = np.atleast_1d(r)  # Pastikan r adalah array.
        mass = self.M_b(r_arr, Sigma_be, a_b)  # Hitung massa.
        v_sq = np.zeros_like(r_arr)  # Inisialisasi v².
        mask = r_arr > 0  # Mask untuk r>0.
        v_sq[mask] = G_ASTRO * mass[mask] / r_arr[mask]  # v² = G M / r.
        return v_sq  # Kembalikan v² bulge.

    @staticmethod  # Metode statis.
    def v_disk_sq(r, Sigma_d0, a_d):  # Kuadrat kecepatan disk (Miyamoto-Nagai approximation).
        r_arr = np.atleast_1d(r)  # Pastikan array.
        v_sq = np.zeros_like(r_arr)  # Inisialisasi.
        mask = r_arr > 0  # Mask r>0.
        r_nonzero = r_arr[mask]  # Radius non-nol.
        y = r_nonzero / (2 * a_d)  # Variabel y untuk Bessel.
        bessel_term = iv(0, y) * kv(0, y) - iv(1, y) * kv(1, y)  # Term Bessel.
        v_sq[mask] = np.pi * G_ASTRO * Sigma_d0 * (r_nonzero**2 / a_d) * bessel_term  # Hitung v² disk.
        return v_sq  # Kembalikan.

    @staticmethod  # Metode statis.
    def v_gas_sq(r):  # Kuadrat kecepatan gas (diasumsikan nol untuk simplifikasi).
        return np.zeros_like(np.atleast_1d(r))  # Selalu nol.

    @staticmethod  # Metode statis.
    def v_nfw_sq(r, rho_0, h):  # Kuadrat kecepatan halo NFW.
        r_arr = np.atleast_1d(r)  # Array.
        v_sq = np.zeros_like(r_arr)  # Inisialisasi.
        mask = r_arr > 0  # Mask.
        r_nonzero = r_arr[mask]  # Non-nol.
        x = r_nonzero / h  # Skala x.
        log_term = np.log(1 + x)  # Log term.
        fraction = x / (1 + x)  # Fraksi.
        v_sq[mask] = 4 * np.pi * G_ASTRO * rho_0 * h**2 * (log_term - fraction) / x  # Formula NFW.
        return v_sq  # Kembalikan.

    @staticmethod  # Metode statis.
    def v_core_sq(r, rho_0, h):  # Kuadrat kecepatan halo core.
        r_arr = np.atleast_1d(r)  # Array.
        v_sq = np.zeros_like(r_arr)  # Inisialisasi.
        mask = r_arr > 0  # Mask.
        r_nonzero = r_arr[mask]  # Non-nol.
        r3_ratio = (r_nonzero / h)**3  # Ratio kubik.
        log_term = np.log1p(r3_ratio)  # Log(1 + ratio).
        v_sq[mask] = (4 * np.pi * G_ASTRO / 3) * rho_0 * (h**3 / r_nonzero) * log_term  # Formula core.
        return v_sq  # Kembalikan.

    def v_baryon_sq(self, r, Sigma_be, a_b, Sigma_d0, a_d):  # Kuadrat kecepatan total baryon (bulge + disk).
        return self.v_bulge_sq(r, Sigma_be, a_b) + self.v_disk_sq(r, Sigma_d0, a_d)  # Jumlahkan.

    def v_mond_std_sq(self, r, v_b_sq):  # Kuadrat kecepatan MOND standar.
        r_arr = np.atleast_1d(r)  # Array r.
        v_b_sq_arr = np.atleast_1d(v_b_sq)  # Array v_b².
        v_m_sq = np.zeros_like(r_arr)  # Inisialisasi.
        mask = v_b_sq_arr > 1e-9  # Mask nilai valid.
        v_b_sq_valid = v_b_sq_arr[mask]  # Valid.
        r_valid = r_arr[mask]  # Valid r.
        term_in_sqrt = 1 + (2 * A0_STD_ASTRO * r_valid / v_b_sq_valid)**2  # Term di akar.
        v_m_sq[mask] = (v_b_sq_valid / np.sqrt(2)) * np.sqrt(1 + np.sqrt(term_in_sqrt))  # Formula MOND std.
        return v_m_sq  # Kembalikan.

    def solve_a_m_arctan(self, a_b, tol=1e-6, verbose=False):  # Solver Newton-Raphson untuk a_M di MOND arctan.
        def f(u):  # Fungsi f(u) = 0 untuk disolve.
            return u * (2 / np.pi) * np.arctan((np.pi * u) / 2) - a_b / A0_ARCTAN_ASTRO
        def f_prime(u):  # Turunan f untuk Newton.
            return (2 / np.pi) * np.arctan((np.pi * u) / 2) + u / (1 + ((np.pi * u) / 2)**2)
        u_guess = a_b / A0_ARCTAN_ASTRO + np.sqrt(a_b / A0_ARCTAN_ASTRO)  # Tebakan awal.
        u = u_guess  # Mulai dari tebakan.
        if verbose:  # Jika verbose, print proses.
            print("\nIterasi Newton-Raphson untuk a_M (arctan):")
            print(f"Estimasi awal u: {u:.6e}")
        converged = False  # Flag konvergensi.
        for i in range(50):  # Maks 50 iterasi.
            delta = f(u) / f_prime(u)  # Hitung delta.
            if verbose:  # Print iterasi.
                print(f"Iterasi {i+1}: u = {u:.6e}, delta = {delta:.6e}")
            u -= delta  # Update u.
            if abs(delta) < tol:  # Cek toleransi.
                if verbose:
                    print(f"Konvergen setelah {i+1} iterasi.")
                converged = True
                break
        if not converged:  # Jika tidak konvergen, warn.
            warnings.warn(f"\nPeringatan: Iterasi Newton-Raphson tidak konvergen setelah 50 iterasi untuk a_b = {a_b:.6e}. "
                         f"Menggunakan nilai terakhir u = {u:.6e} dengan error = {abs(delta):.6e} (tol = {tol}).")
        a_m = u * A0_ARCTAN_ASTRO  # Hitung a_m.
        if verbose:  # Print akhir.
            print(f"a_M akhir: {a_m:.6e} (dalam satuan astro)")
        return a_m  # Kembalikan a_m.

    def v_mond_arctan_sq(self, r, v_b_sq):  # Kuadrat kecepatan MOND arctan.
        r_arr = np.atleast_1d(r)  # Array r.
        v_b_sq_arr = np.atleast_1d(v_b_sq)  # Array v_b².
        v_m_sq = np.zeros_like(r_arr)  # Inisialisasi.
        mask = v_b_sq_arr > 1e-9  # Mask valid.
        for i in np.where(mask)[0]:  # Loop atas valid.
            a_b = v_b_sq_arr[i] / r_arr[i]  # Hitung a_b.
            a_m = self.solve_a_m_arctan(a_b)  # Solve a_m.
            v_m_sq[i] = a_m * r_arr[i]  # v² = a_m * r.
        return v_m_sq  # Kembalikan.

    def v_total_mond_std(self, r, Sigma_be, a_b, Sigma_d0, a_d):  # Kecepatan total MOND std.
        v_b_sq = self.v_baryon_sq(r, Sigma_be, a_b, Sigma_d0, a_d)  # v_b².
        v_total_sq = self.v_mond_std_sq(r, v_b_sq)  # v_total².
        return np.sqrt(np.maximum(0, v_total_sq))  # Akar, hindari negatif.

    def v_total_mond_arctan(self, r, Sigma_be, a_b, Sigma_d0, a_d):  # Kecepatan total MOND arctan.
        v_b_sq = self.v_baryon_sq(r, Sigma_be, a_b, Sigma_d0, a_d)  # v_b².
        v_total_sq = self.v_mond_arctan_sq(r, v_b_sq)  # v_total².
        return np.sqrt(np.maximum(0, v_total_sq))  # Akar.

    def v_total_nfw(self, r, Sigma_be, a_b, Sigma_d0, a_d, rho_0, h):  # Kecepatan total NFW.
        v_b2 = self.v_bulge_sq(r, Sigma_be, a_b)  # Bulge.
        v_d2 = self.v_disk_sq(r, Sigma_d0, a_d)  # Disk.
        v_g2 = self.v_gas_sq(r)  # Gas (nol).
        v_h2 = self.v_nfw_sq(r, rho_0, h)  # Halo.
        v_total_sq = v_b2 + v_d2 + v_g2 + v_h2  # Total².
        return np.sqrt(np.maximum(0, v_total_sq))  # Akar.

    def v_total_core(self, r, Sigma_be, a_b, Sigma_d0, a_d, rho_0, h):  # Kecepatan total core.
        v_b2 = self.v_bulge_sq(r, Sigma_be, a_b)  # Bulge.
        v_d2 = self.v_disk_sq(r, Sigma_d0, a_d)  # Disk.
        v_g2 = self.v_gas_sq(r)  # Gas.
        v_h2 = self.v_core_sq(r, rho_0, h)  # Halo.
        v_total_sq = v_b2 + v_d2 + v_g2 + v_h2  # Total².
        return np.sqrt(np.maximum(0, v_total_sq))  # Akar.

    def set_initial_parameters(self):  # Set parameter awal dan batas untuk fitting per galaksi/model.
        return {  # Dictionary nested: galaksi -> model -> p0 dan bounds.
            'Bima Sakti': {
                'MOND Std': {'p0': [0.66e9, 1.60, 0.61e9, 8.23], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59], [0.70e9, 1.69, 0.96e9, 8.87])},
                'MOND Arctan': {'p0': [0.66e9, 1.60, 0.61e9, 8.23], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59], [0.70e9, 1.69, 0.96e9, 8.87])},
                'ΛCDM NFW': {'p0': [0.66e9, 1.60, 0.61e9, 8.23, 0.71e4, 1.07e3], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59, 0.37e4, 0.73e3], [0.70e9, 1.69, 0.96e9, 8.87, 1.05e4, 1.41e3])},
                'ΛCDM Core': {'p0': [0.66e9, 1.60, 0.61e9, 8.23, 0.71e4, 1.07e3], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59, 0.37e4, 0.73e3], [0.70e9, 1.69, 0.96e9, 8.87, 1.05e4, 1.41e3])}
            },
            'M31': {
                'MOND Std': {'p0': [0.70e9, 1.31, 0.61e9, 4.52], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63], [0.85e9, 1.62, 0.93e9, 6.41])},
                'MOND Arctan': {'p0': [0.70e9, 1.31, 0.61e9, 4.52], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63], [0.85e9, 1.62, 0.93e9, 6.41])},
                'ΛCDM NFW': {'p0': [0.70e9, 1.31, 0.61e9, 4.52, 1.70e7, 14.8], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63, 0.25e7, 7.6], [0.85e9, 1.62, 0.93e9, 6.41, 3.15e7, 22.0])},
                'ΛCDM Core': {'p0': [0.70e9, 1.31, 0.61e9, 4.52, 1.70e7, 14.8], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63, 0.25e7, 7.6], [0.85e9, 1.62, 0.93e9, 6.41, 3.15e7, 22.0])}
            }
        }

    def analyze_model(self, galaxy_name, model_name):  # Analisis fitting untuk model tertentu pada galaksi.
        data = self.data[galaxy_name]  # Ambil data galaksi.
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']  # Ekstrak r, v, sigma.
        fit_function = self.models[model_name]  # Ambil fungsi fitting.
        p0 = self.initial_params[galaxy_name][model_name]['p0']  # Parameter awal.
        bounds = self.initial_params[galaxy_name][model_name]['bounds']  # Batas parameter.

        print(f"\n{'='*80}")  # Print header.
        print(f"ANALISIS: Model {model_name} untuk Galaksi {galaxy_name}")  # Judul analisis.
        print(f"{'='*80}")

        try:  # Coba fitting.
            params, _ = curve_fit(fit_function, r_data, v_data, p0=p0, sigma=sigma_v_data,
                                           bounds=bounds, maxfev=15000, method='trf')  # Lakukan curve_fit.
            print("[SUCCESS] Fitting berhasil.")  # Sukses.

            if 'MOND' in model_name:  # Untuk model MOND.
                Sigma_be, a_b, Sigma_d0, a_d = params  # Unpack params.
                print("\n--- PARAMETER FITTING (MOND) ---")  # Header params.
                print(f"Sigma_be (M_sun/kpc^2): {Sigma_be:.2e}")  # Print Sigma_be.
                print(f"a_b (kpc): {a_b:.2f}")  # Print a_b.
                print(f"Sigma_d0 (M_sun/kpc^2): {Sigma_d0:.2e}")  # Print Sigma_d0.
                print(f"a_d (kpc): {a_d:.2f}")  # Print a_d.
                if model_name == 'MOND Arctan':  # Khusus arctan.
                    print("\n--- NILAI a_M ARCTANGENT DARI METODE ITERASI (untuk 5 radius sampel) ---")  # Header sampel.
                    sample_indices = np.linspace(0, len(r_data)-1, 5, dtype=int)  # Pilih 5 sampel.
                    v_b_sq = self.v_baryon_sq(r_data, Sigma_be, a_b, Sigma_d0, a_d)  # Hitung v_b².
                    for idx in sample_indices:  # Loop sampel.
                        r_sample = r_data[idx]  # Radius sampel.
                        a_b_sample = v_b_sq[idx] / r_sample  # a_b sampel.
                        a_m_sample = self.solve_a_m_arctan(a_b_sample, verbose=True)  # Solve dengan verbose.
                        print(f"Radius {r_sample:.3f} kpc: a_b = {a_b_sample:.6e}, a_M = {a_m_sample:.6e}")  # Print.
            else:  # Untuk ΛCDM.
                Sigma_be, a_b, Sigma_d0, a_d, rho_0, h = params  # Unpack.
                halo_type = 'NFW' if 'NFW' in model_name else 'Core'  # Tipe halo.
                print(f"\n--- PARAMETER FITTING (ΛCDM {halo_type}) ---")  # Header.
                print(f"Sigma_be (M_sun/kpc^2): {Sigma_be:.2e}")  # Print.
                print(f"a_b (kpc): {a_b:.2f}")  # Print.
                print(f"Sigma_d0 (M_sun/kpc^2): {Sigma_d0:.2e}")  # Print.
                print(f"a_d (kpc): {a_d:.2f}")  # Print.
                print(f"rho_0 (M_sun/kpc^3): {rho_0:.2e}")  # Print rho_0.
                print(f"h (kpc): {h:.2f}")  # Print h.

            self.plot_fit_components(galaxy_name, model_name, params, data)  # Plot komponen fit.
            return params  # Kembalikan hasil.

        except (RuntimeError, ValueError) as e:  # Tangkap error fitting.
            print(f"[ERROR] Fitting gagal: {e}")  # Print error.
            return None  # Kembalikan None.

    def plot_fit_components(self, galaxy_name, model_name, params, data):  # Plot fitting dan komponen.
        r_data = data['r']  # Radius data.
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)  # Radius untuk plot (log).
        fit_function = self.models[model_name]  # Fungsi fit.
        v_fit = fit_function(r_plot, *params)  # Hitung v fit.

        plt.figure(figsize=(12, 8))  # Buat figure.
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Data Observasi', capsize=3, markersize=5, zorder=5)  # Plot data dengan error.

        if 'MOND' in model_name:  # Untuk MOND.
            Sigma_be, a_b, Sigma_d0, a_d = params  # Unpack.
            v_bulge_fit = np.sqrt(self.v_bulge_sq(r_plot, Sigma_be, a_b))  # v bulge.
            v_disk_fit = np.sqrt(self.v_disk_sq(r_plot, Sigma_d0, a_d))  # v disk.
            v_baryon_fit = np.sqrt(self.v_baryon_sq(r_plot, Sigma_be, a_b, Sigma_d0, a_d))  # v baryon.
            label = 'MOND Std' if model_name == 'MOND Std' else 'MOND Arctan'  # Label.
            plt.plot(r_plot, v_fit, '-', color='red', linewidth=2.5, label=f'Prediksi Total {label}')  # Plot total.
            plt.plot(r_plot, v_baryon_fit, '--', color='blue', label='Total Baryonik')  # Plot baryon.
            plt.plot(r_plot, v_bulge_fit, ':', color='orange', label='Bulge')  # Plot bulge.
            plt.plot(r_plot, v_disk_fit, ':', color='green', label='Disk')  # Plot disk.
        else:  # Untuk ΛCDM.
            Sigma_be, a_b, Sigma_d0, a_d, rho_0, h = params  # Unpack.
            halo_type = 'NFW' if 'NFW' in model_name else 'Core'  # Tipe.
            v_bulge_fit = np.sqrt(self.v_bulge_sq(r_plot, Sigma_be, a_b))  # v bulge.
            v_disk_fit = np.sqrt(self.v_disk_sq(r_plot, Sigma_d0, a_d))  # v disk.
            v_halo_func = self.v_nfw_sq if 'NFW' in model_name else self.v_core_sq  # Fungsi halo.
            v_halo_fit = np.sqrt(v_halo_func(r_plot, rho_0, h))  # v halo.
            plt.plot(r_plot, v_fit, '-', color='red', linewidth=2.5, label=f'Prediksi Total ΛCDM {halo_type}')  # Total.
            plt.plot(r_plot, v_bulge_fit, '--', color='orange', label='Bulge')  # Bulge.
            plt.plot(r_plot, v_disk_fit, '--', color='green', label='Disk')  # Disk.
            plt.plot(r_plot, v_halo_fit, '--', color='blue', label=f'Halo {halo_type}')  # Halo.

        plt.xscale('log')  # Skala x log.
        plt.xlabel('Radius (kpc)', fontsize=14)  # Label x.
        plt.ylabel('Kecepatan Rotasi (km/s)', fontsize=14)  # Label y.
        plt.title(f'Kurva Rotasi {galaxy_name} - Model {model_name}', fontsize=16)  # Judul.
        plt.legend(loc='best', fontsize=12)  # Legenda.
        plt.grid(True, which="both", ls="--", alpha=0.5)  # Grid.
        plt.ylim(bottom=0)  # Y min 0.
        plt.tight_layout()  # Layout rapi.
        plt.show()  # Tampilkan plot.

    @staticmethod  # Metode statis.
    def calculate_statistics(v_obs, sigma_v, v_model, k):  # Hitung statistik fitting.
        residuals = v_obs - v_model  # Residual.
        chi_sq = np.sum((residuals / sigma_v) ** 2)  # Chi-square.
        n = len(v_obs)  # Jumlah data.
        dof = n - k  # Degrees of freedom.
        chi_sq_reduced = chi_sq / dof if dof > 0 else float('inf')  # Reduced chi².
        aic = 2 * k + chi_sq  # AIC.
        bic = np.log(n) * k + chi_sq  # BIC.
        r_squared = r2_score(v_obs, v_model)  # R².
        return chi_sq, chi_sq_reduced, aic, bic, r_squared  # Kembalikan.

    def compare_model_statistics(self, galaxy, results):  # Bandingkan statistik model.
        print(f"\n{'='*80}")  # Header.
        print(f"EVALUASI STATISTIK UNTUK {galaxy.upper()}")  # Judul.
        print(f"{'='*80}")
        print(f"{'Model':<20} | {'χ²':>10} | {'χ²/dof':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10}")  # Tabel header.
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}")  # Garis pemisah.

        data = self.data[galaxy]  # Data galaksi.
        v_data, sigma_v_data = data['v'], data['sigma_v']  # v dan sigma.
        model_stats = {}  # Dict stats.

        for model_name, result in results.items():  # Loop model.
            if result['params'] is not None:  # Jika sukses.
                k = len(result['params'])  # Jumlah params.
                v_model = result['fit_function'](data['r'], *result['params'])  # Hitung v model.
                chi2, chi2_red, aic, bic, r2 = self.calculate_statistics(v_data, sigma_v_data, v_model, k)  # Stats.
                model_stats[model_name] = {'aic': aic, 'bic': bic}  # Simpan AIC/BIC.
                print(f"{model_name:<20} | {chi2:>10.2f} | {chi2_red:>10.2f} | {aic:>10.2f} | {bic:>10.2f} | {r2:>10.4f}")  # Print baris.

        print(f"{'-'*80}\n")  # Pemisah.

        if 'MOND Std' in model_stats and 'MOND Arctan' in model_stats:  # Bandingkan MOND.
            delta_aic = model_stats['MOND Arctan']['aic'] - model_stats['MOND Std']['aic']  # Delta AIC.
            print(f"Perbandingan AIC (Arctan - Std): {delta_aic:.2f}")  # Print.
            if delta_aic > 10:  # Interpretasi.
                print("-> Bukti sangat kuat mendukung MOND Std.")
            elif delta_aic > 2:
                print("-> Bukti positif mendukung MOND Std.")
            elif delta_aic < -10:
                print("-> Bukti sangat kuat mendukung MOND Arctan.")
            elif delta_aic < -2:
                print("-> Bukti positif mendukung MOND Arctan.")
            else:
                print("-> Model setara.")

        if 'ΛCDM NFW' in model_stats and 'ΛCDM Core' in model_stats:  # Bandingkan ΛCDM.
            delta_aic = model_stats['ΛCDM Core']['aic'] - model_stats['ΛCDM NFW']['aic']  # Delta.
            print(f"Perbandingan AIC (Core - NFW): {delta_aic:.2f}")  # Print.
            if delta_aic > 10:  # Interpretasi.
                print("-> Bukti sangat kuat mendukung ΛCDM NFW.")
            elif delta_aic > 2:
                print("-> Bukti positif mendukung ΛCDM NFW.")
            elif delta_aic < -10:
                print("-> Bukti sangat kuat mendukung ΛCDM Core.")
            elif delta_aic < -2:
                print("-> Bukti positif mendukung ΛCDM Core.")
            else:
                print("-> Model setara.")

    def compare_all_models_statistics(self, galaxy, results):  # Perbandingan komprehensif semua model.
        print(f"\n{'='*80}")  # Header.
        print(f"PERBANDINGAN KOMPREHENSIF SEMUA MODEL - {galaxy.upper()}")  # Judul.
        print(f"{'='*80}")

        data = self.data[galaxy]  # Data.
        v_data, sigma_v_data = data['v'], data['sigma_v']  # v, sigma.

        model_stats = {}  # Dict stats.
        valid_models = {}  # Dict model valid.

        for model_name, result in results.items():  # Loop.
            if result['params'] is not None:  # Valid.
                k = len(result['params'])  # k.
                v_model = result['fit_function'](data['r'], *result['params'])  # v model.
                chi2, chi2_red, aic, bic, r2 = self.calculate_statistics(v_data, sigma_v_data, v_model, k)  # Stats.
                model_stats[model_name] = {  # Simpan.
                    'chi2': chi2, 'chi2_red': chi2_red, 'aic': aic, 'bic': bic, 'r2': r2
                }
                valid_models[model_name] = True  # Tandai valid.

        if not valid_models:  # Jika tidak ada.
            print("Tidak ada model yang valid untuk perbandingan.")  # Print.
            return

        print(f"{'Model':<20} | {'χ²':>10} | {'χ²/dof':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10} | {'Rank AIC':>8} | {'Rank BIC':>8}")  # Header tabel.
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*10}|{'-'*10}")  # Pemisah.

        sorted_aic = sorted(model_stats.items(), key=lambda x: x[1]['aic'])  # Sort AIC.
        sorted_bic = sorted(model_stats.items(), key=lambda x: x[1]['bic'])  # Sort BIC.

        aic_ranks = {model: i+1 for i, (model, _) in enumerate(sorted_aic)}  # Rank AIC.
        bic_ranks = {model: i+1 for i, (model, _) in enumerate(sorted_bic)}  # Rank BIC.

        for model_name, stats in model_stats.items():  # Loop print baris.
            print(f"{model_name:<20} | {stats['chi2']:>10.2f} | {stats['chi2_red']:>10.2f} | "
                  f"{stats['aic']:>10.2f} | {stats['bic']:>10.2f} | {stats['r2']:>10.4f} | "
                  f"{aic_ranks[model_name]:>8} | {bic_ranks[model_name]:>8}")

        best_aic_model, best_aic_stats = sorted_aic[0]  # Model terbaik AIC.
        best_bic_model, best_bic_stats = sorted_bic[0]  # Terbaik BIC.

        print(f"\n--- HASIL PERBANDINGAN MODEL TERBAIK ---")  # Header hasil.
        print(f"Berdasarkan AIC: {best_aic_model}")  # Print.
        print(f"  - AIC: {best_aic_stats['aic']:.2f}")  # AIC.
        print(f"  - ΔAIC terkecil: Model ini memiliki AIC terendah")  # Info.

        print(f"Berdasarkan BIC: {best_bic_model}")  # Print.
        print(f"  - BIC: {best_bic_stats['bic']:.2f}")  # BIC.
        print(f"  - ΔBIC terkecil: Model ini memiliki BIC terendah")  # Info.

        print(f"\n--- PERBANDINGAN RELATIF TERHADAP MODEL TERBAIK ---")  # Header relatif.
        print(f"{'Model':<20} | {'ΔAIC':>10} | {'ΔBIC':>10} | {'Kekuatan Bukti':>15}")  # Tabel.
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*17}")  # Pemisah.

        for model_name, stats in model_stats.items():  # Loop.
            delta_aic = stats['aic'] - best_aic_stats['aic']  # Delta AIC.
            delta_bic = stats['bic'] - best_bic_stats['bic']  # Delta BIC.

            if delta_aic == 0:  # Interpretasi bukti.
                evidence = "TERBAIK"
            elif delta_aic <= 2:
                evidence = "Setara"
            elif delta_aic <= 6:
                evidence = "Sedang"
            elif delta_aic <= 10:
                evidence = "Lemah"
            else:
                evidence = "Tidak mendukung"

            print(f"{model_name:<20} | {delta_aic:>10.2f} | {delta_bic:>10.2f} | {evidence:>15}")  # Print baris.

        print(f"\n--- ANALISIS CHI-SQUARE REDUCED ---")  # Header chi².
        chi2_red_sorted = sorted(model_stats.items(), key=lambda x: abs(x[1]['chi2_red'] - 1))  # Sort chi² red.
        best_chi2_red_model, best_chi2_red_stats = chi2_red_sorted[0]  # Terbaik.

        print(f"Model dengan χ²/dof terdekat ke 1: {best_chi2_red_model}")  # Print.
        print(f"  - χ²/dof: {best_chi2_red_stats['chi2_red']:.2f}")  # Nilai.

        for model_name, stats in model_stats.items():  # Loop kualitas.
            deviation = abs(stats['chi2_red'] - 1)  # Deviasi.
            quality = "Sangat baik" if deviation <= 0.5 else "Baik" if deviation <= 1.0 else "Cukup" if deviation <= 2.0 else "Kurang"  # Kualitas.
            print(f"  {model_name:<18}: {stats['chi2_red']:.2f} ({quality})")  # Print.

        print(f"\n--- REKOMENDASI MODEL ---")  # Header rekomendasi.

        if best_aic_model == best_bic_model:  # Jika sama.
            print(f"✓ MODEL TERBAIK: {best_aic_model}")  # Print.
            print(f"  - Konsisten berdasarkan AIC dan BIC")  # Info.
            print(f"  - AIC: {best_aic_stats['aic']:.2f}, BIC: {best_aic_stats['bic']:.2f}")  # Nilai.
            print(f"  - χ²/dof: {best_aic_stats['chi2_red']:.2f}, R²: {best_aic_stats['r2']:.4f}")  # Stats.
        else:  # Jika berbeda.
            print(f"✓ REKOMENDASI BERDASARKAN AIC: {best_aic_model}")  # AIC.
            print(f"  - AIC: {best_aic_stats['aic']:.2f}, ΔAIC: +0.00")  # Nilai.
            print(f"  - χ²/dof: {best_aic_stats['chi2_red']:.2f}, R²: {best_aic_stats['r2']:.4f}")  # Stats.

            print(f"✓ REKOMENDASI BERDASARKAN BIC: {best_bic_model}")  # BIC.
            print(f"  - BIC: {best_bic_stats['bic']:.2f}, ΔBIC: +0.00")  # Nilai.
            print(f"  - χ²/dof: {best_bic_stats['chi2_red']:.2f}, R²: {best_bic_stats['r2']:.4f}")  # Stats.

            aic_params = len(results[best_aic_model]['params'])  # Jumlah params AIC.
            bic_params = len(results[best_bic_model]['params'])  # BIC.

            if bic_params < aic_params:  # Saran berdasarkan parsimony.
                print(f"✓ SARAN: {best_bic_model} lebih parsimonious (lebih sedikit parameter)")
            else:
                print(f"✓ SARAN: {best_aic_model} memberikan fit lebih baik dengan parameter tambahan yang justified")

        print(f"{'='*80}\n")  # Akhir.

    def plot_model_comparison(self, galaxy, results):  # Plot perbandingan NFW vs Core dan MOND Std vs Arctan.
        data = self.data[galaxy]  # Data.
        r_data = data['r']  # r.
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)  # r plot.

        plt.figure(figsize=(12, 8))  # Figure ΛCDM.
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Observasi', capsize=3, markersize=5)  # Data.

        if 'ΛCDM NFW' in results and results['ΛCDM NFW']['params'] is not None:  # Jika valid.
            v_nfw = self.v_total_nfw(r_plot, *results['ΛCDM NFW']['params'])  # v NFW.
            plt.plot(r_plot, v_nfw, 'b--', label='ΛCDM NFW')  # Plot.

        if 'ΛCDM Core' in results and results['ΛCDM Core']['params'] is not None:  # Valid.
            v_core = self.v_total_core(r_plot, *results['ΛCDM Core']['params'])  # v core.
            plt.plot(r_plot, v_core, 'g:', label='ΛCDM Core')  # Plot.

        plt.xscale('log')  # Log x.
        plt.xlabel('Radius (kpc)')  # Label.
        plt.ylabel('Kecepatan Rotasi (km/s)')  # Label.
        plt.title(f'Perbandingan ΛCDM NFW vs Core untuk {galaxy}')  # Judul.
        plt.legend()  # Legenda.
        plt.grid(True, which='both')  # Grid.
        plt.ylim(bottom=0)  # Ymin.
        plt.show()  # Tampilkan.

        plt.figure(figsize=(12, 8))  # Figure MOND.
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Observasi', capsize=3, markersize=5)  # Data.

        if 'MOND Std' in results and results['MOND Std']['params'] is not None:  # Valid.
            v_std = self.v_total_mond_std(r_plot, *results['MOND Std']['params'])  # v std.
            plt.plot(r_plot, v_std, 'r-', label='MOND Std')  # Plot.

        if 'MOND Arctan' in results and results['MOND Arctan']['params'] is not None:  # Valid.
            v_arctan = self.v_total_mond_arctan(r_plot, *results['MOND Arctan']['params'])  # v arctan.
            plt.plot(r_plot, v_arctan, 'm-.', label='MOND Arctan')  # Plot.

        plt.xscale('log')  # Log.
        plt.xlabel('Radius (kpc)')  # Label.
        plt.ylabel('Kecepatan Rotasi (km/s)')  # Label.
        plt.title(f'Perbandingan MOND Std vs Arctan untuk {galaxy}')  # Judul.
        plt.legend()  # Legenda.
        plt.grid(True, which='both')  # Grid.
        plt.ylim(bottom=0)  # Ymin.
        plt.show()  # Tampilkan.

    def plot_all_models_comparison(self, galaxy, results):  # Plot semua model bersama.
        data = self.data[galaxy]  # Data.
        r_data = data['r']  # r.
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)  # r plot.

        plt.figure(figsize=(12, 8))  # Figure.
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Data Observasi', capsize=3, markersize=5)  # Data.

        styles = {  # Style per model.
            'MOND Std': {'color': 'red', 'linestyle': '-'},
            'MOND Arctan': {'color': 'magenta', 'linestyle': '-.'},
            'ΛCDM NFW': {'color': 'blue', 'linestyle': '--'},
            'ΛCDM Core': {'color': 'green', 'linestyle': ':'}
        }

        for model_name, result in results.items():  # Loop.
            if result['params'] is not None:  # Valid.
                v_fit = result['fit_function'](r_plot, *result['params'])  # v fit.
                plt.plot(r_plot, v_fit, label=model_name, **styles[model_name])  # Plot.

        plt.xscale('log')  # Log.
        plt.xlabel('Radius (kpc)', fontsize=14)  # Label.
        plt.ylabel('Kecepatan Rotasi (km/s)', fontsize=14)  # Label.
        plt.title(f'Perbandingan Semua Model untuk {galaxy}', fontsize=16)  # Judul.
        plt.legend(loc='best', fontsize=12)  # Legenda.
        plt.grid(True, which="both", ls="--", alpha=0.5)  # Grid.
        plt.ylim(bottom=0)  # Ymin.
        plt.tight_layout()  # Rapi.
        plt.show()  # Tampilkan.

    def plot_residuals(self, galaxy, results):  # Plot residual per model.
        data = self.data[galaxy]  # Data.
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']  # Ekstrak.

        fig, axes = plt.subplots(len(results), 1, figsize=(12, 10), sharex=True)  # Subplots vertikal.
        if len(results) == 1:  # Jika satu, jadikan list.
            axes = [axes]
        fig.suptitle(f'Residual Analisis untuk {galaxy}', fontsize=16)  # Suptitle.

        colors = {'MOND Std': 'red', 'MOND Arctan': 'magenta', 'ΛCDM NFW': 'blue', 'ΛCDM Core': 'green'}  # Warna.

        for i, (model_name, result) in enumerate(results.items()):  # Loop model.
            if result['params'] is not None:  # Valid.
                v_model = result['fit_function'](r_data, *result['params'])  # v model.
                residuals = v_data - v_model  # Residual.
                axes[i].errorbar(r_data, residuals, yerr=sigma_v_data, fmt='o',
                                color=colors[model_name], ecolor='lightgray', capsize=3)  # Plot residual dengan error.
                axes[i].axhline(0, color='black', linestyle='--')  # Garis nol.
                axes[i].set_ylabel('Residual (km/s)')  # Label.
                axes[i].set_title(model_name)  # Judul sub.
                axes[i].grid(True)  # Grid.

        plt.xscale('log')  # Log x.
        plt.xlabel('Radius (kpc)')  # Label x.
        plt.tight_layout()  # Rapi.
        plt.show()  # Tampilkan.

    def compare_models(self, galaxy, model1, model2, results):  # Perbandingan dua model.
        data = self.data[galaxy]  # Data.
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']  # Ekstrak.

        if model1 not in results or model2 not in results:  # Cek ada.
            print(f"Data tidak tersedia untuk {model1} vs {model2}")  # Print.
            return

        params1 = results[model1]['params']  # Params1.
        params2 = results[model2]['params']  # Params2.
        if params1 is None or params2 is None:  # Cek valid.
            print(f"Fitting gagal untuk {model1} atau {model2}. Perbandingan dilewati.")  # Print.
            return

        func1 = results[model1]['fit_function']  # Fungsi1.
        func2 = results[model2]['fit_function']  # Fungsi2.

        v_pred1 = func1(r_data, *params1)  # v pred1.
        v_pred2 = func2(r_data, *params2)  # v pred2.

        k1 = len(params1)  # k1.
        k2 = len(params2)  # k2.

        chi_sq1, _, aic1, bic1, r2_1 = self.calculate_statistics(v_data, sigma_v_data, v_pred1, k1)  # Stats1.
        chi_sq2, _, aic2, bic2, r2_2 = self.calculate_statistics(v_data, sigma_v_data, v_pred2, k2)  # Stats2.

        delta_chi_sq = chi_sq1 - chi_sq2  # Delta chi².
        delta_aic = aic1 - aic2  # Delta AIC.
        delta_bic = bic1 - bic2  # Delta BIC.

        print(f"\n{'='*80}")  # Header.
        print(f"PERBANDINGAN {galaxy.upper()}: {model1} vs {model2}")  # Judul.
        print(f"{'='*80}")
        print(f"{'Model':<20} | {'χ²':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10}")  # Tabel.
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}")  # Pemisah.

        print(f"{model1:<20} | {chi_sq1:>10.2f} | {aic1:>10.2f} | {bic1:>10.2f} | {r2_1:>10.4f}")  # Baris1.
        print(f"{model2:<20} | {chi_sq2:>10.2f} | {aic2:>10.2f} | {bic2:>10.2f} | {r2_2:>10.4f}")  # Baris2.

        print("\nPerbedaan:")  # Header delta.
        print(f"Δχ² ({model1} - {model2}): {delta_chi_sq:.2f}")  # Print.
        print(f"ΔAIC ({model1} - {model2}): {delta_aic:.2f}")  # Print.
        print(f"ΔBIC ({model1} - {model2}): {delta_bic:.2f}")  # Print.

        print("\nInterpretasi Berdasarkan AIC:")  # Header interpretasi.
        if delta_aic > 10:  # Kondisi.
            print(f"  Bukti sangat kuat {model2} lebih baik.")
        elif delta_aic > 6:
            print(f"  Bukti kuat {model2} lebih baik.")
        elif delta_aic > 2:
            print(f"  Bukti sedang {model2} lebih baik.")
        elif delta_aic > 0:
            print(f"  Bukti lemah {model2} lebih baik.")
        elif delta_aic < -10:
            print(f"  Bukti sangat kuat {model1} lebih baik.")
        elif delta_aic < -6:
            print(f"  Bukti kuat {model1} lebih baik.")
        elif delta_aic < -2:
            print(f"  Bukti sedang {model1} lebih baik.")
        elif delta_aic < 0:
            print(f"  Bukti lemah {model1} lebih baik.")
        else:
            print("  Kedua model setara berdasarkan AIC.")

        df = abs(k1 - k2)  # df untuk chi² test.
        p_value = stats.chi2.sf(abs(delta_chi_sq), df) if df > 0 else 1.0  # P-value.

        print(f"\nUji Signifikansi Chi-kuadrat:")  # Header uji.
        print(f"P-value: {p_value:.4f}")  # Print.

        if p_value < 0.05:  # Interpretasi.
            better_model = model1 if delta_chi_sq < 0 else model2
            print(f"  Perbedaan signifikan (p < 0.05). Model terbaik: {better_model}")
        else:
            print("  Tidak ada perbedaan signifikan (p ≥ 0.05)")

        print(f"{'='*80}\n")  # Akhir.

    def run_analysis(self):  # Jalankan analisis keseluruhan.
        print("Memulai analisis kurva rotasi galaksi...")  # Mulai.
        all_results = {}  # Dict hasil semua galaksi.

        for galaxy in ['Bima Sakti', 'M31']:  # Loop galaksi.
            print(f"\n{'#' * 50}")  # Header galaksi.
            print(f" MEMPROSES GALAKSI: {galaxy.upper()} ")  # Judul.
            print(f"{'#' * 50}")

            galaxy_results = {}  # Hasil per galaksi.
            for model_name in self.models.keys():  # Loop model.
                params = self.analyze_model(galaxy, model_name)  # Analisis.
                galaxy_results[model_name] = {  # Simpan.
                    'params': params,
                    'fit_function': self.models[model_name]
                }

            all_results[galaxy] = galaxy_results  # Simpan ke all.

            self.compare_models(galaxy, 'ΛCDM NFW', 'MOND Std', galaxy_results)  # Bandingkan pasang-pasang.
            self.compare_models(galaxy, 'ΛCDM Core', 'MOND Std', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM NFW', 'ΛCDM Core', galaxy_results)
            self.compare_models(galaxy, 'MOND Std', 'MOND Arctan', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM NFW', 'MOND Arctan', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM Core', 'MOND Arctan', galaxy_results)

            if any(res['params'] is not None for res in galaxy_results.values()):  # Jika ada valid.
                print(f"\n--- VISUALISASI DAN STATISTIK UNTUK {galaxy.upper()} ---")  # Header vis.
                self.compare_model_statistics(galaxy, galaxy_results)  # Stat sederhana.

                self.compare_all_models_statistics(galaxy, galaxy_results)  # Komprehensif.

                self.plot_model_comparison(galaxy, galaxy_results) # Plot perbandingan NFW vs Core dan MOND Std vs Arctan.

                self.plot_all_models_comparison(galaxy, galaxy_results)  # Semua model.

                self.plot_residuals(galaxy, galaxy_results) # Plot residual

            else:  # Tidak ada.
                print(f"\n[INFO] Tidak ada model yang berhasil difitting untuk {galaxy}. Visualisasi dilewati.")  # Info.

        print("\nAnalisis selesai!")  # Akhir.

if __name__ == "__main__":  # Jalankan jika skrip utama.
    analyzer = GalaxyRotationAnalyzer()  # Buat objek.
    analyzer.run_analysis()  # Jalankan analisis.
```
